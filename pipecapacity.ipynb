```python
import math
```

```python

def PipeCapacity(Q, n, S):
    if S <= 0:
        raise ValueError("S0 must be greater than zero")
    S0 = 1 / S
    numerator = 3.21 * Q * n
    denominator = math.sqrt(S0)
    result = (numerator / denominator) ** (3 / 8)
    result_mm = result*1000
    return result_mm, round(result_mm)
```

```python
try:
    Q = float(input("Flow rate Q (in m³/s): "))
    n = float(input("Manning's roughness coefficient n: "))
    S0 = float(input("Bed slope S0: "))

    exact_D, rounded_D = PipeCapacity(Q, n, S0)
    print(f"\nComputed diameter D: {exact_D:.2f} mm")
    print(f"Rounded diameter D: {rounded_D} mm")

except ValueError as e:
    print(f"Invalid input: {e}")
```
